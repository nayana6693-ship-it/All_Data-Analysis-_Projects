# SuperStore ETL & EDA Project

Using Pandas, NumPy, Matplotlib and Seaborn

```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize']=(10,5)
sns.set_style('whitegrid')

df = pd.read_csv('Sample - Superstore(2).csv', encoding='windows-1252')
df.head()
```

## Data Understanding & Cleaning

```

df.info()
df.describe()
df.isnull().sum()
df.duplicated().sum()
```

```

df['Cost'] = ((df['Sales']-(df['Sales']*df['Discount']))-df['Profit'])
df['Product_Price'] = df['Sales']/df['Quantity']

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df['Delivery Time'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Order_Month'] = df['Order Date'].dt.month_name()
df['Order_Day'] = df['Order Date'].dt.day_name()

df.head()
```

## Exploratory Data Analysis

```

# Most Frequent Product
top_products = df['Product Name'].value_counts().head(10)
top_products.plot(kind='bar', title='Top 10 Sold Products')
plt.show()
```

```

# Region, State, City Orders
fig, ax = plt.subplots(1,3, figsize=(18,5))

df['Region'].value_counts().plot(kind='bar', ax=ax[0], title='Region')
df['State'].value_counts().head(15).plot(kind='bar', ax=ax[1], title='Top States')
df['City'].value_counts().head(15).plot(kind='bar', ax=ax[2], title='Top Cities')

plt.tight_layout()
plt.show()
```

```

# Discounts
(df['Discount']*100).value_counts().sort_index().plot(kind='bar')
plt.title('Discount Distribution')
plt.show()
```

```

# Category and Sub-Category
df['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Category Distribution')
plt.show()

df['Sub-Category'].value_counts().plot(kind='barh')
plt.show()
```

```

# Loyal Customers
df['Customer Name'].value_counts().head(10).plot(kind='bar')
plt.title('Top Loyal Customers')
plt.show()
```

```

# Business Flourish Day & Month
fig, ax = plt.subplots(1,2, figsize=(16,5))

sns.countplot(data=df, x='Order_Day',
              order=df['Order_Day'].value_counts().index, ax=ax[0])

sns.countplot(data=df, x='Order_Month',
              order=df['Order_Month'].value_counts().index, ax=ax[1])

ax[1].tick_params(axis='x', rotation=90)
plt.tight_layout()
plt.show()
```

## Sales & Profit Analysis

```

sales_profit = df.groupby('Customer Name')[['Sales','Profit']].sum()
sales_profit['Ratio'] = sales_profit['Profit'] / sales_profit['Sales']
sales_profit.sort_values('Sales', ascending=False).head(10)
```

```

region_sales = df.groupby('Region')[['Sales','Profit']].sum()
region_sales.plot(kind='bar')
plt.title('Region Sales & Profit')
plt.show()
```

```

segment_profit = df.groupby('Segment')['Profit'].sum().sort_values(ascending=False)
segment_profit.plot(kind='pie', autopct='%1.1f%%')
plt.title('Profit by Segment')
plt.show()
```

```

top_product_sales = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_product_profit = df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)

print(top_product_sales)
print(top_product_profit)
```

```

best_category = df.groupby('Category')[['Sales','Profit']].sum()
best_category.plot(kind='bar')
plt.show()
```

## Bivariate Analysis

```

# Profitable Category by Region
pivot1 = pd.pivot_table(df, values='Profit', index='Region', columns='Category', aggfunc='sum')
sns.heatmap(pivot1, annot=True, fmt='.0f')
plt.show()
```

```

# Bestseller Day by Category
pd.crosstab(df['Order_Day'], df['Category']).plot(kind='bar', figsize=(10,5))
plt.show()
```

```

# Profitable Ship Mode by Region
pivot2 = pd.pivot_table(df, values='Profit', index='Region', columns='Ship Mode', aggfunc='sum')
pivot2.plot(kind='bar')
plt.show()
```

```

# Bestseller Month by Category
pd.crosstab(df['Order_Month'], df['Category']).plot(kind='bar', figsize=(12,5))
plt.xticks(rotation=90)
plt.show()
```

```

# Bestseller SubCategory by Segment
pd.crosstab(df['Sub-Category'], df['Segment']).plot(kind='barh', figsize=(10,8))
plt.show()
```

## Multivariate Analysis

```

corr = df[['Sales','Quantity','Discount','Profit','Cost','Product_Price','Delivery Time']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Analysis')
plt.show()
```
